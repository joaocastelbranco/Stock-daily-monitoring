```python
#import
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# DEfine make graph function
def make_graph(stock_data1,stock_data2,stock_data3):
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price Apple", "Historical Share Price Tesla", "Historical Share Google"), vertical_spacing = .3)
    stock_data1_specific = stock_data1[stock_data1.Date <= '2022-06-27']
    stock_data2_specific = stock_data2[stock_data2.Date <= '2022-06-27']
    stock_data3_specific = stock_data3[stock_data3.Date <= '2022-06-27']
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data1_specific.Date, infer_datetime_format=True), y=stock_data1_specific.Close.astype("float"), name="Share Price1"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data2_specific.Date, infer_datetime_format=True), y=stock_data2_specific.Close.astype("float"), name="Share Price2"), row=2, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data2_specific.Date, infer_datetime_format=True), y=stock_data3_specific.Close.astype("float"), name="Share Price3"), row=3, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_xaxes(title_text="Date", row=3, col=1)    
    fig.update_yaxes(title_text="Price USD", row=1, col=1)
    fig.update_yaxes(title_text="Price USD", row=2, col=1)
    fig.update_yaxes(title_text="Price USD", row=3, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title="stock",
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
# Get Tesla data
tesla = yf.Ticker("TSLA")
tesla_data=tesla.history(period="max")
# Reset index
tesla_data.reset_index(inplace=True)
tesla_data.tail()
```

```python
# Get Google data
google = yf.Ticker("GOOG")
google_data=google.history(period="max")
# Reset index
google_data.reset_index(inplace=True)
google_data.tail()
```

```python
# Get Apple data
apple = yf.Ticker("AAPL")
apple_data=apple.history(period="max")
# Reset index
apple_data.reset_index(inplace=True)
apple_data.tail()
```

```python
make_graph(apple_data,tesla_data,google_data)
```
